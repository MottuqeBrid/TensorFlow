```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("data/insurance_data.csv")
data.head()
```

```python
data.shape
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()
```

```python
x = data[["age"]]
y = data["bought_insurance"]
x, y
```

```python
model.fit(x, y)
```

```python
model.coef_
```

```python
model.intercept_
```

```python
prediction = model.predict(x)
prediction
```

```python
fig = plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="age", y="bought_insurance")
plt.plot(x, prediction, color="red", marker="*", linestyle="--")
plt.plot(x.mean(), y.mean(), color="green", marker="o", markersize=10, label="Mean")
plt.title("Age vs Bought Insurance")
```
